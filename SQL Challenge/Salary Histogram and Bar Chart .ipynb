```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
```

```python
engine = create_engine('postgresql+psycopg2://postgres:postgres@localhost:5432/employee_db')
connection = engine.connect()
```

```python
salaries = pd.read_sql("Select * from salary", connection)
```

```python
salary = salaries['salary']
plt.hist(salary, bins=6)
plt.title('Histogram of Salaries')
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.style.use('ggplot')
```

```python
avg_sal = pd.read_sql("Select * from avg_sals", connection)
```

```python
plt.figure(figsize=(10,5))
plt.bar(avg_sal['title'], avg_sal['avg_salary'])
plt.title('Average Salaries By Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.xticks(rotation='45')
plt.show()
```
